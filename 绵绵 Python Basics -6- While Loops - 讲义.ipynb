# While Loops

- 当while statement的条件满足时（the condition is true）
- while loop 会一遍一遍的run自己内部的code，直到条件不满足时（the condition is no longer met）

```python
num = 0

while num < 5:
    print(num)
    num += 1
print('finished')
```

```python
num = 0 

while num <= 10:
    if num % 2 == 0:
        print(f'{num} is an even number')
    else:
        print(f'{num} is an odd number')
    num += 1
```

```python
# 在使用while loops的时候要注意更新条件里的参数，不然会陷入无限循环

num = 0

while num < 5:
    print(num)
```

### Keywords: break, continue 

- 在loops中使用（一般和if statement一起）
- break: 走出当前的loop
- continue: 直接进入下一个循环

#### break

```python
num = 6 

while True:
    print(num)
    num += 1
    
    if num % 5 == 0:
        break
print('out of the while loop')
```

```python
s = 'hello world'

for char in s:
    print(char)
    if char == 'r':
        break
print('out of the for loop')
```

#### continue

```python
lst = [1, 1, 2, 3, 3, 4, 5, 6, 7]

for num in lst:
    if num == 3 or num == 6:
        print("don't print this number")
    else:
        print(num)
```

```python
lst = [1, 1, 2, 3, 3, 4, 5, 6, 7]

for num in lst:
    if num == 3 or num == 6:
        continue
    print(num)
```
